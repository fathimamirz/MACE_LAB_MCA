```python
import pandas as pd
```

```python
fpath='/content/drive/MyDrive/DSML-LAB/iris.csv'
df=pd.read_csv(fpath)
```

```python
df.head(10)
```

```python
import matplotlib.pyplot as plt
```

```python
col='sepal.length'
plt.hist(df[col],edgecolor='red')
plt.xlabel(col)
plt.ylabel('frequency')
plt.show()
```

```python
import seaborn as sns
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='variety',y=col,data=df)
plt.xlabel('column for quaratile plot')
plt.ylabel('frequency')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df[col],kde=True)
plt.xlabel('column for distribution')
plt.ylabel('density')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='sepal.length', y='sepal.width', hue='variety', data=df)
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.title('Sepal Length vs Sepal Width by Species')
plt.legend(title='Species', loc='upper right')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='sepal.length', y='sepal.width', hue='variety', style='variety', data=df, s=100)
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.title('Sepal Length vs Sepal Width by Species')
plt.legend(title='Species', loc='upper right')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.pairplot(df,hue="variety",diag_kind="kde",palette="viridis")
plt.suptitle("distribution chart",y=1.02)
plt.show()
```

```python
ax = df.plot(kind="scatter", x="sepal.length",y="sepal.width", color="b", label="sepal.length vs sepal.width")
df.plot(kind="scatter", x="sepal.length",y="petal.width", color="r", label="sepal.length vs petal.width", ax=ax)
df.plot(kind="scatter", x="sepal.length",y="petal.length", color="g", label="sepal.length vs petal.length", ax=ax)

ax.set_xlabel("sepal.length")
ax.set_ylabel("vertical label")
plt.show()
```

```python
sns.pairplot(df, hue='variety', markers=['o', 's', 'D'], diag_kind='kde', height=2.5)
plt.suptitle('Scatter Matrix of Iris Dataset by Species', y=1.02)
plt.show()
```

```python
x = df['sepal.length']
y = df['sepal.width']
size =df['petal.length'] * 10
plt.figure(figsize=(10, 6))
plt.scatter(x, y, s=size, alpha=0.7, c='red', edgecolors='black')

plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.title('Bubble Chart - Sepal Length vs Sepal Width (Petal Length as Bubble Size)')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.kdeplot(df['sepal.length'], shade=True, color='blue')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Density')
plt.title('Density Plot - Sepal Length')
plt.show()
```

```python

from pandas.plotting import parallel_coordinates
plt.figure(figsize=(10, 6))
parallel_coordinates(df, 'variety', colormap='viridis')

plt.legend(title='variety', loc='upper right')

plt.title('Parallel Coordinates Plot - Iris Dataset')
plt.show()
```

```python
mean_sepal_length = df['sepal.length'].mean()

df['Deviation'] =df['sepal.length'] - mean_sepal_length

plt.figure(figsize=(10, 6))
plt.bar(df['variety'], df['Deviation'], color='red')

plt.xlabel('Species')
plt.ylabel('Deviation from Mean Sepal Length')
plt.title('Deviation Chart - Sepal Length')

plt.show()
```

```python
from pandas.plotting import andrews_curves
plt.figure(figsize=(10, 6))
andrews_curves(df, 'variety', colormap='viridis')

plt.legend(title='variety', loc='upper right')

plt.title('Andrews Curves - Iris Dataset')
plt.show()
```
